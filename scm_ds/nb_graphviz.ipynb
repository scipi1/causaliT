```python
from graphviz import Digraph
from scm import *
from scm import *
import os
os.environ["PATH"] += r";C:\Program Files\Graphviz\bin"
```

```python
def dag_graphviz(scm) -> Digraph:
    g = Digraph('SCM', format='pdf', graph_attr={'rankdir':'LR'})
    for v, spec in scm.specs.items():
        g.node(v)
        for p in spec.parents:
            g.edge(p, v)
    return g
```

```python
specs = [
        NodeSpec("X", [],            "eps_X"),                  # exogenous
        NodeSpec("Z", [],            "eps_Z"),                  # exogenous (scaled)
        NodeSpec("Y", ["X", "Z"],    "X + w_YZ*Z + eps_Y"),     # structural
    ]


scm = SCM(specs)
dag_graphviz(scm)
```
